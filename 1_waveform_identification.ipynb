# Waveform Identification

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

from utils import zoll
from utils import misc
from utils.autoscale import autoscale
```

```python
# parameters
case_id = 29
step_id = 1
```

```python
case_str = str(case_id).zfill(5)
zoll_filename = 'zoll' + case_str + '.json'
json = zoll.LoadJSON("data/" + zoll_filename)
```

```python
def zoomed_plot(df, samplerate, start=900, end=910):
    if df is not None:
        duration = len(df) / samplerate
    if df is None or duration < 10:
        print("Nothing to plot.")
        return False
    time = np.arange(0, duration, 1 / samplerate) # time vector

    fig = plt.figure(figsize=(20,4))
    ax = plt.gca()
    plt.xlim(start, end)
    plt.title("10 seconds of waveform")
    plt.plot(time, df['Waveform'].tolist(), color="red")
    autoscale(ax, 'y')
    return True
```

## Channel 0:

```python
df, samplerate = zoll.LoadWaveform(json, hide_starttime=True, channel=0)
```

```python
if df is not None:
    df['Waveform'].plot(figsize=(20,4), title="Channel 0")
    plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
    plt.savefig(f'reports/images/1_{case_str}_01.png', bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

```python
if zoomed_plot(df, samplerate) == True:
    plt.savefig(f'reports/images/1_{case_str}_02.png', bbox_inches='tight', pad_inches=0.1)
    plt.show()
```
